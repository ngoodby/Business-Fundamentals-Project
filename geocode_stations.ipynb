```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import geopy.geocoders
```

```python
df = pd.read_excel('data/EV_Charging_Data.xlsx', sheet_name = 'data')
```

```python
locations = pd.DataFrame(df.Location.unique(), columns = ["Address"])
locations
```

```python
from geopandas.tools import geocode

def custom_geocoder(address):
    dataframe = geocode(address , provider="nominatim" , user_agent = 'my_request')
    point = dataframe.geometry.iloc[0]
    return pd.Series({'Latitude': point.y, 'Longitude': point.x})
 
locations[['latitude' , 'longitude']]= locations.Address.apply( lambda x: custom_geocoder(x))
```
